### Neat Progress

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load fitness log ===
fitness_df = pd.read_csv("neat_logs/neat_training_log.csv")

plt.figure(figsize=(10, 5))
plt.plot(fitness_df["generation"], fitness_df["avg_fitness"], label="Avg Fitness")
plt.plot(fitness_df["generation"], fitness_df["best_fitness"], label="Best Fitness")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.title("NEAT Fitness over Generations")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV with no header
ep_len_df = pd.read_csv("neat_logs/neat_ep_len_log.csv", header=None, names=["episode", "avg_ep_len"])

# Plot
plt.figure(figsize=(10, 5))
plt.plot(ep_len_df["episode"], ep_len_df["avg_ep_len"], marker="o", label="Avg Episode Length")
plt.xlabel("Episode")
plt.ylabel("Avg Episode Length (timesteps)")
plt.title("NEAT Episode Length over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### PPO Progress

```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

# Replace with your PPO log folder path
log_dir = "ppo_logs/PPO_CarRacer_1"

event_acc = EventAccumulator(log_dir)
event_acc.Reload()

# Plot only available key
rew_mean = event_acc.Scalars("rollout/ep_rew_mean")
steps = [x.step for x in rew_mean]
mean_vals = [x.value for x in rew_mean]

plt.figure(figsize=(10, 5))
plt.plot(steps, mean_vals, label="PPO Mean Reward")
plt.xlabel("Timesteps")
plt.ylabel("Reward")
plt.title("PPO Mean Reward")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
ep_len_events = event_acc.Scalars("rollout/ep_len_mean")
steps = [e.step for e in ep_len_events]
ep_lens = [e.value for e in ep_len_events]

# === 4. Plot ===
plt.figure(figsize=(10, 5))
plt.plot(steps, ep_lens, label="PPO Avg Episode Length", color="purple")
plt.xlabel("Timesteps")
plt.ylabel("Episode Length")
plt.title("PPO Episode Length over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
